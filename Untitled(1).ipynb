## IMPORT LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 202431014_Muhamad Fadlan
image_path = 'Diri.jpg'  
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_rgb)
plt.title('Citra Asli')
plt.axis('off')
plt.show()
```

## Median Filtering

```python
# 202431014_Muhamad Fadlan
median_filtered = cv2.medianBlur(img_rgb, 5)

cv2.imwrite('Hasil_Median_Filter.jpg', cv2.cvtColor(median_filtered, cv2.COLOR_RGB2BGR))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1); plt.imshow(img_rgb); plt.title('Citra Asli'); plt.axis('off')
plt.subplot(1, 2, 2); plt.imshow(median_filtered); plt.title('After Filter Median'); plt.axis('off')
plt.show()
```

## Fungsi Mean Filtering

```python
# 202431014_Muhamad Fadlan
def manual_mean_filter(image, kernel_size=5):
    kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size**2)
    
    rows, cols = image.shape
    pad = kernel_size // 2
    padded_img = cv2.copyMakeBorder(image, pad, pad, pad, pad, cv2.BORDER_REFLECT)
    output = np.zeros_like(image, dtype=np.float32)
    
    for i in range(rows):
        for j in range(cols):
            roi = padded_img[i:i+kernel_size, j:j+kernel_size]
            output[i, j] = np.sum(roi * kernel)
            
    return output.astype(np.uint8)
```

## Hasil Mean Filtering

```python
# 202431014_Muhamad Fadlan
mean_filtered = manual_mean_filter(img_gray, 5)

cv2.imwrite('Hasil_Mean_Filter.jpg', mean_filtered)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1); plt.imshow(img_gray, cmap='gray'); plt.title('Original Image'); plt.axis('off')
plt.subplot(1, 2, 2); plt.imshow(mean_filtered, cmap='gray'); plt.title('Mean Filtered Image'); plt.axis('off')
plt.show()
```
